````python
# ‚ñ∂ Zone1 Î∞úÏ†ÑÎüâÍ≥º Ï†ÑÎ†• Î∞úÏÇ∞Îüâ(diffuse flows) ÎπÑÍµê
 #Ïù¥Ï§ëÏ∂ïÏúºÎ°ú ÌôïÏù∏
  #fig, ax1 = plt.subplots() ÏÑ†Ïñ∏
   #ax1.twinx()             ÏßÑÌñâ
import scipy.stats as stats


fig, ax1 = plt.subplots()                       #subplotÏùÑ ÏÇ¨Ïö©ÌïòÏßÄ ÏïäÏïÑÏÑú plt.subplots() Î°ú ÎÅùÎÇ∏ Í≤É
ax1.plot(df['DateTime'], df['Zone 1 Power Consumption'], color='green', label='Zone1')
ax1.axhline(df['Zone 1 Power Consumption'].mean(), color = 'green')

ax2 = ax1.twinx()
ax2.plot(df['DateTime'], df['diffuse flows'], color='deeppink', label='Diffuse')
ax2.axhline(df['diffuse flows'].mean(), color = 'deeppink')


fig.legend()                                       #Ïù¥Ï§ëÏ∂ïÏù¥Í∏∞ ÎïåÎ¨∏Ïóê plt.legend() Í∞Ä ÏïÑÎãå fig.legend()
plt.gcf().set_size_inches(25, 5)
print(stats.pearsonr(df['Zone 1 Power Consumption'], df['diffuse flows']))

plt.show()
````
